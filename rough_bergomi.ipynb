### Case1: different values of eta - call option

```python
H = 0.1
eta = np.linspace(0.1, 1.5, 10)
X0 = np.log(0.235**2)
T = [1 / 12, 3 / 12, 6 / 12]
kappa = 0.2
N = 300
M = 10**5
```

```python
rbergomi = MonteCarloRoughBergomi(eta=eta[0], H=H, X0=X0, T=T[0], kappa=kappa, N=N, M=M)
```

```python
rbergomi.future()
```

```python
VIX_mc_1m = 100 * np.array(
    [
        MonteCarloRoughBergomi(eta=e, H=H, X0=X0, T=T[0], kappa=kappa, N=N, M=M).call()
        for e in eta
    ]
)
```

```python
VIX_approx_1m = 100 * np.array(
    [WeakApproxRoughBergomi(e, H, X0, T[0], kappa).call() for e in eta]
)
```

```python
from utils import relative_error


rel_error_1m = relative_error(VIX_approx_1m, VIX_mc_1m)
```
